### Target Marketing

Target marketing involves breaking a market into segments and then concentrating your marketing efforts on one or a few key segments consisting of the customers whose needs and desires most closely match your product or service offerings. It can be the key to attracting new business, increasing sales, and making your business a success
[[1]](https://www.thebalancesmb.com/target-marketing-2948355).

### Market Segmentation
Market segmentation is the process of dividing a target market into smaller, more defined categories. It segments customers and audiences into groups that share similar characteristics such as demographics, interests, needs, or location.
The four bases of market segmentation are [[2]](https://blog.alexa.com/types-of-market-segmentation/#:~:text=Market%20segmentation%20is%20the%20process,interests%2C%20needs%2C%20or%20location.):
* Demographic segmentation
* Psychographic segmentation
* Behavioral segmentation
* Geographic segmentation

**Market Segmentation is crucial for maximizing market campaign conversion rate**

* CREDIT_LIMIT : Limit of Credit Card for user

* MINIMUM_PAYMENTS : Minimum amount of payments made by user
* PRC_FULL_PAYMENT : Percent of full payment paid by user

# Data Preprocessing


```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

```python
df_scaled.shape
```

# K-Means
## Finding optimal value of 'k' i.e no. of clusters using Elbow Method

```python
from sklearn.cluster import KMeans
```

```python
wcss = [] # within cluster squared sum

k_range = range(1, 21)

for k in k_range:
    
    print("No. of Clusters:",k)
    kmeans = KMeans(n_clusters=k)
    
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)
    print("WCSS =",wcss[k-1])
    print("----------------------------------------------------------")

plt.plot(k_range, wcss, color="b")
plt.scatter(k_range, wcss, color="r")
plt.show()
```

```python
# we choose k=4
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(df_scaled)
labels = kmeans.labels_
```

```python
kmeans.cluster_centers_.shape # centroid of clusters
```

```python
cluster_centers = pd.DataFrame(data = kmeans.cluster_centers_, columns = [df.columns])
cluster_centers = scaler.inverse_transform(cluster_centers)
cluster_centers = pd.DataFrame(data = cluster_centers, columns = [df.columns])
cluster_centers
```

```python
# concatenate the clusters labels to our original dataframe
df_cluster = pd.concat([df, pd.DataFrame({'cluster':labels})], axis = 1)
df_cluster.head()

# Plot the histogram of various clusters
for i in df.columns:

    plt.figure(figsize = (35, 5))

    for j in range(kmeans.cluster_centers_.shape[0]):

        plt.subplot(1,kmeans.cluster_centers_.shape[0],j+1)

        cluster = df_cluster[df_cluster['cluster'] == j]

        cluster[i].hist(bins = 20)

        plt.title('{}    \nCluster {} '.format(i,j))
  
    plt.show()

# This plot will show you distribution of data in a cluster w.r.t a column
```

# PCA for visualization of clusters

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2) # 2 if we want 2d representation

pca_comp = pca.fit_transform(df_scaled)

print(pca.explained_variance_ratio_)
```

```python
pca_comp
```

```python
# Create a dataframe with the two components
pca_df = pd.DataFrame(data = pca_comp, columns =['pca1','pca2'])
pca_df.head()

# Concatenate the clusters labels to the dataframe
pca_df = pd.concat([pca_df,pd.DataFrame({'cluster':labels})], axis = 1)
pca_df.head()

plt.figure(figsize=(10,10))
ax = sns.scatterplot(x="pca1", y="pca2", hue = "cluster", data = pca_df, palette =['red','green','blue','pink'])#'yellow','gray','purple', 'black'])
plt.show()
```
